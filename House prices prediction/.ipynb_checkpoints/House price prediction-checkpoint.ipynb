## 1. Introduction:



**Forecasting Problem**

We are attempting to predict the Airbnb listing prices, based on attributes like availability, popularity and location features using the diverse dataset shared. The forecasting model has real life implications for different parties, such as for Airbnb hosts they can factor in the implications to optimise occupancy,pricing strategy, improving and optimising services provided to the occupants and also to understand trends that would help make their property more lucrative to occupants. Guests can use it to better plan their trips and get their value for money.Alternatively, Airbnb can use it to implement policy changes and match customers with properties using an effective strategic approach.

**Evaluation Criteria**

Each evaluation task has an individual performance metric.The criteria is to asses forecast performance that will depend on the specific task. For price, which is a continuous variable prediction, and other variables like number of reviews, metrics like RMSE and MAPE can prove to be more effective. For binary outcomes like listing availability, Accuracy, Precision, Recall or F1 score can be more effective.We are using the validation data to test our model and evaluate the performance using RMSE,R2,Adjusted R2, MPE and MAPE.


**Types of Variables/Features**

The dataset includes a collection of categorical, numerical, boolean and temporal variables. For categorical we have qualitative variables that can be divided into multiple categories but having no order or priority. For example, 'room_type' is categorical but 'Entire home/apt', 'Private room','host_location' and 'neighbourhood' are also categorical features. They are handled by one-hot encoding for machine learning algorithms. 


Numerical are quantifiable variables and can be either discrete (e.g number of reviews) or continuous (e.g latitude and longitude).

Boolean variables represent two possible states: True or False(eq 'host_has_profile_pic'). These variables are useful in differentiating between two states, like whether a host is a superhost or not, which could influence the popularity or price of a listing.

These diverse features provide comprehensive insights into the multifaceted Airbnb listing prediction problem.

```python
# Convert date columns to datetime format
date_columns = ['host_since', 'first_review', 'last_review']
for col in date_columns:
     df_train[col] = pd.to_datetime(df_train[col])
```

```python
# Get the most recent date in 'host_since' , Assuming it is the cut off date
latest_date_in_data = df_train['last_review'].max()

# Print the most recent date
print(f"The most recent date in the 'last_review' column is: {latest_date_in_data}")

# Create 'host_days' to record the number of days the host has been on the platform
for column in date_columns:
    df_train[column] = pd.to_datetime(df_train[column])
    df_train['days_' + column] = (latest_date_in_data - df_train[column]).dt.days
    df_test[column] = pd.to_datetime(df_test[column])
    df_test['days_' + column] = (latest_date_in_data - df_test[column]).dt.days
```

### Creating new feature from `amenities`

The amenities column contains a list of additional features in the property, i.e. whether it has a TV or parking. For example, a property might have the following amenities: "TV", "Wifi", "Air conditioning", "Parking", "Balcony" etc.

We will create new features from this column. For each amenity, we will create a new column in the DataFrame. For example, for the 'TV' amenity, we will create a new column called 'TV'. If a property has a TV, the value in this column will be 1, else it will be 0.

To achieve this, we first need to create a set of all possible amenities.

```python
# Example of amenities listed
df_train.amenities[:1].values
```

```python
# Creating a set of all possible amenities
# List of the relevant datasets
datasets = [df_train, df_test]

# Perform operation on each dataset
for dataset in datasets:
    amenities_list = list(dataset.amenities)
    amenities_list_string = " ".join(amenities_list)
    amenities_list_string = amenities_list_string.replace('[', '')
    amenities_list_string = amenities_list_string.replace(']', ',')
    amenities_list_string = amenities_list_string.replace('"', '')
    amenities_set = [x.strip() for x in amenities_list_string.split(',')]
    amenities_set = set(amenities_set)
amenities_set 
```

Certain amenities hold more significance than others in terms of their impact on price, such as a balcony being more likely to increase the price compared to a fax machine. Additionally, there are certain amenities that are less commonly found, like an electric profiling bed.

The block of code down is mapping amenities in the dataset to more simplified feature names, represented in a dictionary. This is done to convert these categorical amenities into binary features. Each amenity in a listing is checked against this dictionary, creating new binary columns in the dataframe, indicating the presence (1) or absence (0) of that amenity in a listing. This process is done for both the training and test datasets.

### Creating feature from the geographical data
Geographical data such as latitude and longitude can be a crucial component when predicting property prices, as prices can greatly vary depending on the location of a property. By creating clusters based on geographical data, we can divide the area into different segments that share similar characteristics, such as being close to the city center or in the suburbs.

**Visualizing Geographical Data**

To start with, we plot a scatter plot of the geographical data, where each point represents a property and its geographical location.

In this plot, you can see a rough outline of Melbourne, and the distribution of listings resembles the map of the city. Each dot represents a property listed on Airbnb, and the overall pattern gives an idea of the density and spread of the listings.

```python
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
# Scatter plot of the geographical data
plt.figure(figsize=(10, 8))
plt.scatter(df_train['longitude'], df_train['latitude'], alpha=0.2)
plt.title('Geographical data')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

# Find optimal number of clusters
sil = [] # silhouette score placeholder
K = range(30, 50) # iterate from 2 to 15 clusters
for k in K:
    kmeans = KMeans(n_clusters=k).fit(df_train[['latitude', 'longitude']])
    preds = kmeans.predict(df_train[['latitude', 'longitude']])
    sil.append(silhouette_score(df_train[['latitude', 'longitude']], preds))

# Plot K against Silhouette score to find optimal number of clusters
plt.figure(figsize=(10, 8))
plt.plot(K, sil, 'bx-')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Optimal k')
plt.show()

# Define the maximum number of clusters you want to check for
max_clusters = 50

# Prepare an empty list to store the inertia for each number of clusters
inertia = []

# Calculate inertia for different numbers of clusters
for k in range(1, max_clusters + 1):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_train[['latitude', 'longitude']])
    inertia.append(kmeans.inertia_)

# Plot the results
plt.figure(figsize=(10, 8))
plt.plot(range(1, max_clusters + 1), inertia, 'bx-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.show()
```

**Determining the Optimal Number of Clusters**
To create clusters based on the geographical data, we use the KMeans algorithm from the sklearn library. To determine the optimal number of clusters, we use the silhouette score.

First, I calculate the silhouette score for a range of number of clusters (here we use 30 to 50). The silhouette score is a measure of how similar an object is to its own cluster compared to other clusters. A higher silhouette score indicates a better-defined cluster structure.But, we will go for the one perfoms better on test data which **37** weidly close to number of suburbs.

Next, we calculate the inertia for a range of number of clusters. Inertia measures the total distance of data points from their respective cluster centers. We look for the "elbow" in the plot where the decrease in inertia slows significantly. This gives us a good estimate of the optimal number of clusters.

**Clustering Geographical Data**
After determining the optimal number of clusters, we create a KMeans object and fit it to our latitude and longitude data. The result is a prediction of the cluster for each observation.

```python
# Create a KMeans object
kmeans = KMeans(n_clusters=37, random_state=0)

# Fit the KMeans object to the latitude and longitude columns and predict the cluster for each observation
clusters = kmeans.fit_predict(df_train[['latitude', 'longitude']])

# Add the clusters to the dataframe
df_train['cluster'] = clusters

# Do the same for the test set
clusters_test = kmeans.predict(df_test[['latitude', 'longitude']])
df_test['cluster'] = clusters_test
```

**Visualizing the Clusters**

Next, we create a scatter plot of the clusters. Each color represents a different cluster, allowing us to visualize the distribution of the clusters geographically.

```python
# Create a scatter plot of the data color-coded by cluster
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df_train, x='longitude', y='latitude', hue='cluster', palette='viridis')
plt.title('Clusters of Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

**Creating Dummy Variables for the Clusters**

The clustering created a categorical variable, and categorical variables are often more useful in a format of binary (dummy) variables. Therefore, we need to convert our newly created 'cluster' column to dummy variables, so our machine learning model can use it effectively.

```python
# Create dummy variables for the clusters
# Concatenate these dummy variables to the original dataframe
# Note: the prefix 'c' is added to each dummy variable
df_train = pd.concat([df_train, pd.get_dummies(df_train['cluster'], prefix='c')], axis=1)
df_test = pd.concat([df_test, pd.get_dummies(df_test['cluster'], prefix='c')], axis=1)

# List of calculated host listings count columns
clustercorr = ['price','cluster']

ccorr = df_train[clustercorr].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(ccorr, square=True, cmap="RdYlGn_r", annot=True)
```

Here above, pd.get_dummies() is used to convert the 'cluster' categorical variable into dummy/indicator variables. Each cluster is given its own column in the dataframe, where it is marked with a '1' if the property belongs to that cluster, and '0' if it does not. The prefix 'c' is added to each dummy variable to make the column names more descriptive.

### Imputing missing values:

### Imputing missing values  for `bathrooms`, `bedrooms` , `beds`, `days_host_since`, `days_first_review` and `days_last_review`

```python
# Impute missing values with the median
for col in ['bathrooms', 'bedrooms','beds','days_host_since', 'days_first_review', 'days_last_review']:
    df_train[col].fillna(df_train[col].median(), inplace=True)
    df_test[col].fillna(df_test[col].median(), inplace=True)
```

## 3. Fit and tune a forecasting model/Submit predictions/Report score and ranking


**Retaining Numerical Features**

First, we retain only the numerical columns from the training and test data. This is because most machine learning models can only handle numerical inputs. Therefore, we need to convert our categorical variables into numbers.

```python
# for training data
df_train = df_train.select_dtypes(include=np.number)
# for testing data
df_test = df_test.select_dtypes(include=np.number)
```

**Storing and Removing ID Column**

We store the 'ID' columns in separate series and then remove them from the datasets because 'ID' is not a feature that can contribute to the predictions. The 'price' column is also dropped from the test dataset as it is the target variable and not included in the test set.

```python
# Store the 'ID' columns in separate series
train_ids = df_train['ID']
test_ids = df_test['ID']

# Remove the 'ID' columns from the dataframes
df_train= df_train.drop(columns=['ID'])
df_test= df_test.drop(columns=['ID'])

df_test.drop(['price'], axis=1, inplace=True)
```

**Importing Machine Learning Libraries**

We then import a range of libraries that will be used to create different machine learning models. These libraries include models like Linear Regression, Decision Tree, Random Forest, Gradient Boosting, and others.

```python
from sklearn.model_selection import train_test_split, GridSearchCV,KFold, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.metrics import mean_squared_error
from sklearn.svm import SVR
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor
from numpy import sqrt
```

### Model Selection Without Hyperparameter Tuning
We first run multiple models without hyperparameter tuning. This gives us a baseline model with default parameters to compare our tuned models against. We define our target variable as 'price' and use all other numerical variables as features. We then split our training data into a training set and a validation set.

We define a dictionary of different models and fit each model on the training data, and evaluate it on the validation data. The performance metric used here is the Root Mean Square Error (RMSE), which gives us an absolute measure on how our models predictions deviate from the actual values.

```python
# Define your target variable (what you want to predict)
target = 'price'

# Define your features (all other columns that are not the target)
features = [col for col in df_train.columns if col != target]

# Split your training data into training and validation set
X_train, X_val, y_train, y_val = train_test_split(df_train[features],
                                                  df_train[target], test_size=0.1, random_state=42)

# Create a dictionary of different models
models = {'Linear Regression': LinearRegression(), 
          'Decision Tree': DecisionTreeRegressor(random_state=42),
          'Random Forest': RandomForestRegressor(random_state=42),
          'Gradient Boosting': GradientBoostingRegressor(random_state=42),
          'Support Vector Machines': SVR(),
          'XGBoost': XGBRegressor(random_state=42),
          'LightGBM': LGBMRegressor(random_state=42),
          'AdaBoost': AdaBoostRegressor(random_state=42)}

# Initialize a dictionary to hold your results
results = {}
# Fit each model and evaluate it
for model_name, model in models.items():
    model.fit(X_train, y_train)
    predictions = model.predict(X_val)
    results[model_name] = sqrt(mean_squared_error(y_val, predictions))

# Determine which model was the best
best_model_name = min(results, key=results.get)
best_model = models[best_model_name]

print(f'The best model was {best_model_name} with RMSE {results[best_model_name]}')

# Now fit the best model to all of the training data
best_model.fit(df_train[features], df_train[target])
```

### Hyperparameter Tuning Using GridSearchCV
To potentially improve our model, we perform hyperparameter tuning on the best model identified in the previous step. We use GridSearchCV, which exhaustively tries every combination of provided hyperparameters to find the best model.

The model's performance, however, did not significantly improve. This could be due to various reasons such as the model's limitations in capturing complex patterns, or the hyperparameters selected for tuning did not include the optimal values.

```python
# Define your target variable (what you want to predict)
target = 'price'

# Define your features (all other columns that are not the target)
features = [col for col in df_train.columns if col != target]

# Create the model and parameters dictionary
model = GradientBoostingRegressor(random_state=42)
params = {'n_estimators': [80, 100, 120], 'learning_rate': [0.05, 0.1, 0.15]}

# Initialize the GridSearchCV object
grid_search = GridSearchCV(model, params, cv=3, scoring='neg_mean_squared_error', n_jobs=-1)

# Fit GridSearchCV
grid_search.fit(df_train[features], df_train[target])

# Get the best model
best_model = grid_search.best_estimator_

# Perform cross-validation and calculate the RMSE
scores = cross_val_score(best_model, df_train[features], df_train[target], 
                         scoring='neg_mean_squared_error', cv=10, n_jobs=-1)
rmse_scores = [sqrt(-x) for x in scores]
avg_rmse = sum(rmse_scores) / len(rmse_scores)

print(f'The best Gradient Boosting model has an average RMSE (from CV): {avg_rmse}')

# Now fit the best model to all of the training data
best_model.fit(df_train[features], df_train[target])
```

#### Generating and Submitting Predictions
Finally, we impute missing values in the test set using the mean strategy, generate predictions using the best model, and prepare a submission file.


```python
from sklearn.impute import SimpleImputer

# Create an instance of the imputer with mean strategy
imputer = SimpleImputer(strategy='mean')

# Fit the imputer on the training data and transform the test data
imputed_test_data = imputer.fit_transform(df_test)

# Make predictions on the imputed test data using the best_model
price_predictions = best_model.predict(imputed_test_data)

# Create a DataFrame for the submission
submission = pd.DataFrame({'ID': test_ids, 'price': price_predictions})

# Write the submission DataFrame to a CSV file
submission.to_csv('submission.csv', index=False)
```
